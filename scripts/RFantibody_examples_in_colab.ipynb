```python
%%bash
export PYTHONPATH=$PYTHONPATH:/content/RFantibody/src/rfantibody/rfdiffusion
cd /content/RFantibody/src/rfantibody/rfdiffusion/
poetry run python /content/RFantibody/scripts/rfdiffusion_inference.py \
    --config-path /content/RFantibody/src/rfantibody/rfdiffusion/config/inference/ \
    --config-name antibody \
    antibody.target_pdb=/content/RFantibody/scripts/examples/example_inputs/rsv_site3.pdb \
    antibody.framework_pdb=/content/RFantibody/scripts/examples/example_inputs/hu-4D5-8_Fv.pdb \
    inference.ckpt_override_path=/content/RFantibody/weights/RFdiffusion_Ab.pt \
    'ppi.hotspot_res=[T305,T456]' \
    'antibody.design_loops=[L1:8-13,L2:7,L3:9-11,H1:7,H2:6,H3:5-13]' \
    inference.num_designs=3 \
    inference.output_prefix=/content/RFantibody/scripts/examples/example_outputs/ab_des
```
